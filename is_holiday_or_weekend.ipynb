```python
import pandas as pd
```

```python
from holidayskr import is_holiday
```

```python
df = pd.read_csv('apple_data.csv',encoding='cp949'); df
```

```python
from datetime import datetime, date

def check_weekend(date):
    weekday = date.weekday()
    if weekday > 4:
        return True
    else:
        return False

check_weekend(datetime.strptime(df.loc[12,'date'],'%Y-%m-%d'))
```

```python
df['holiday'] = df['date'].apply(
    lambda date: 1 if is_holiday(date) or check_weekend(datetime.strptime(date, '%Y-%m-%d')) else 0
)
```

```python
df = df[['date', 'holiday', 'temp', 'temp_range', 'prec', 'mean_wind', 'max_inst_wind',
       'name', 'species', 'product grade', 'price', 'discount', 'etl time']]
```
